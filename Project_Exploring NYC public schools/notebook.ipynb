Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data

#best math schools
schools.head()
topscore=.8*800
best=schools[schools["average_math"]>=topscore]
best_math_schools= best.loc[:,("school_name","average_math")].sort_values(by="average_math", ascending=False)
best_math_schools.head()

#best combined scores
schools.head()
combined_scores=schools[["school_name", "average_math", "average_reading", "average_writing"]]
combined_scores["total_SAT"]=combined_scores["average_math"]+combined_scores["average_reading"]+combined_scores["average_writing"]
combined_scores=combined_scores.sort_values(by='total_SAT',ascending=False)
combined_scores.head()
top_10_schools=combined_scores.iloc[0:10,:]
top_10_schools=top_10_schools.loc[:, ["school_name", "total_SAT"]]
print(top_10_schools)


#largest standard deviation
combined_scores=schools[["school_name","borough","average_math", "average_reading", "average_writing"]]
combined_scores.head()
combined_scores["total_SAT"]=combined_scores["average_math"]+combined_scores["average_reading"]+combined_scores["average_writing"]
results= pd.DataFrame()
results["num_schools"]= combined_scores.groupby("borough").size()
results["average_SAT"]= combined_scores.groupby("borough")["total_SAT"].mean().round(2)
results["std_SAT"]= combined_scores.groupby("borough")["total_SAT"].std().round(2)
results.head()
largest_std_dev=results[results["std_SAT"] == results["std_SAT"].max()]
print(largest_std_dev)

# Start coding here...
# Add as many cells as you like...
```
